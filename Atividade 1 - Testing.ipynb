```python
sns.set(font_scale=0.75)
df5 = df4.corr().round(decimals=2)
mask = np.tril(np.ones_like(df5, dtype=np.bool))
ax = plt.figure(figsize=(10,10))

heatmap = sns.heatmap(df5.round(decimals=2), mask=mask, linewidths=.5,
                      vmin=-1, vmax=1, square=True, annot=True)
```

```python
df10 = df4['blueTotalGold']
df10
```

```python
fig, ax = plt.subplots()

y_pos = np.arange(len(df10.value_counts()[:20]))
values = df10.values[:20]
gameindex = (np.arange(len(df10.value_counts()[:20])) +1)

ax.barh(y_pos, values, align='center', color='blue')
ax.set_yticks(y_pos)
ax.set_yticklabels(gameindex)
ax.invert_yaxis()
ax.set_xlabel('Total de ouro')
ax.set_ylabel('Número do Jogo')
ax.set_title('Total de ouro por jogos')

plt.show()
```

```python
btg = df4[['blueWins', 'blueTotalGold']]
brg = btg.sort_values('blueTotalGold', ascending=True)

#sns.set(font_scale=1.4)
sns.lmplot('blueTotalGold','blueWins', btg)
# \n
plt.title('Blue Total Gold x Results')
plt.xlabel('Result')
plt.ylabel('Blue Total Gold')
```
